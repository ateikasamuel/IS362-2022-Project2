```python
# Import pandas and numpy, DataFrame/Series
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

#Read the contents of Student Test Performance.csv into a DataFrame
Student_file = pd.read_csv("D:/IS362_Project2/Student Test Performance/Student Test Performance.csv")

#Display the contents of Student Test Performance.csv DataFrame
Student_file
```

```python
# Tidying the data:
# Split the dataset into two Dataframes named "Student_info & Student_Grades"

# Create a new DataFrame
student_info = Student_file.copy()

# Remove four columns name is 'Test number','Term 1', 'Term 2', and 'Term 3'
student_info.drop(["Test number","Term 1", "Term 2", "Term 3"], axis=1, inplace=True)

# Display the results of student info
student_info
```

```python
# Split the column "Sex and age" and create separate columns for each. 
student_info[['Sex', 'Age']] = student_info['Sex and age'].str.split('_', expand=True)

# Removing the "Sex and age" column
student_info.drop(["Sex and age"], axis=1, inplace=True)

# Display the result of the DataFrame
student_info
```

```python
# Remove the duplicate rows for each student
student_info.drop_duplicates(['ID'], keep='first', inplace=True)

# Display the result of the DataFrame
student_info
```

```python
# Calculate the average age of each Student
avg_age = student_info['Age'].median()

# Display the result of the calculation
avg_age
```

```python
# Identify the youngest student's age
youngest = student_info['Age'].min()

# Display the youngest student
youngest
```

```python
# Identify the oldest student's age
oldest = student_info['Age'].max()

# Display the oldest student
oldest
```

```python
# Count the number of male and female in the class
count_by_sex = student_info["Sex"].value_counts()

# Display the count of females and males in the class
count_by_sex
```

```python
# Create a new DataFrame
student_grades = Student_file.copy()

# Remove four columns name is 'Name", "Phone", and "Sex and age"
student_grades.drop(["Name", "Phone", "Sex and age"], axis=1, inplace=True)

# Display the results of student info
student_grades
```

```python
# Transpose Term 1, Term 2, Term 3 to a new column named Term and the associated grades to a new column named Score
grades = pd.melt(student_grades, id_vars=['ID', 'Test number'], var_name='Term', value_name='Score')

# Sort the grades by Term
grades.sort_values(by=['Term'], inplace=True)

#Display the grades
grades
```

```python
# Calculate the average score of each test, for each term
avg_test = grades.groupby(["Test number", "Term"])["Score"].mean()

# Display the average score of each test, for each term
avg_test
```

```python
# Calculate the lowest grade for each test, for each term
lowest = grades.groupby(["Test number", "Term"])["Score"].min()

# Display the lowest grade for each test, for each term
lowest
```

```python
# Calculate the highest grade for each test, for each term
highest = grades.groupby(["Test number", "Term"])["Score"].max()

# Display the highest grade for each test, for each term
highest
```
