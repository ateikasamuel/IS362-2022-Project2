```python
# Import pandas and numpy, DataFrame/Series
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

#Read the contents of combined_sales.csv into a DataFrame
combined_sales = pd.read_csv("D:/IS362_Project2/Stores_Combined_Sales/combined_sales.csv")

#Display the contents of combined_sales.csv DataFrame
combined_sales
```

```python
# Create a new DataFrame
noblesville = combined_sales.copy()

# Remove the columns "Bloomington", "Unnamed: 5", "Muncie", "Unnamed: 7", "South Bend", "Unnamed: 9", "Fort Wayne", "Unnamed: 11"
noblesville.drop(["Bloomington", "Unnamed: 5", "Muncie", "Unnamed: 7", "South Bend", "Unnamed: 9", "Fort Wayne", "Unnamed: 11"], 
                 axis=1, inplace=True)

# Rename columns in the Noblesville dataset
noblesville.rename(columns = {'Unnamed: 0':'Month', 'Unnamed: 1':'Day','Noblesville':'Price','Unnamed: 3':'Quantity'}, 
                   inplace = True)

# Drop the first row that contained the previous labels
noblesville.drop([0], inplace = True )

# Sort Dataframe by Month
noblesville.sort_values(by='Month', inplace=True,)

# Display the results of Noblesville
noblesville
```

```python
# Calculate the quantity each recorded day of each month
nobs_avg = noblesville.groupby(['Month','Day'])['Quantity'].mean()

# Display the quantity each recorded day of each month
nobs_avg
```

```python
# Calculate the lowest quantity for the month
nobs_min = noblesville.groupby(['Month'])['Quantity'].min()

# Display the lowest quantity for the month
nobs_min
```

```python
# Calculate the highest quantity for the month
nobs_min = noblesville.groupby(['Month'])['Quantity'].max()

# Display the highest quantity for the month
nobs_min
```
